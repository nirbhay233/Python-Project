```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("Student.csv")

# Dropping any rows with missing values
df.dropna(inplace=True)

# Converting categorical columns to the appropriate data types
df['Choose your gender'] = df['Choose your gender'].astype('category')
df['Marital status'] = df['Marital status'].astype('category')
df['Do you have Depression?'] = df['Do you have Depression?'].map({'Yes': 1, 'No': 0})
df['Do you have Anxiety?'] = df['Do you have Anxiety?'].map({'Yes': 1, 'No': 0})
df['Do you have Panic attack?'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})
df['Did you seek any specialist for a treatment?'] = df['Did you seek any specialist for a treatment?'].map({'Yes': 1, 'No': 0})

# Checking for duplicates and dropping them
df.drop_duplicates(inplace=True)

# Displaying the first few rows of the cleaned data
print(df.head())
```

```python
# Summary statistics
print(df.describe())

# Distribution of ages
df['Age'].hist(bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Distribution of CGPA
df['What is your CGPA?'].hist(bins=20)
plt.title('CGPA Distribution')
plt.xlabel('CGPA')
plt.ylabel('Frequency')
plt.show()

# Count plot for gender
df['Choose your gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Count plot for marital status
df['Marital status'].value_counts().plot(kind='bar')
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
labels = df['Choose your gender'].value_counts().index
values = df['Choose your gender'].value_counts().values

plt.figure(figsize=(8, 6))
plt.bar(labels, values, color=['blue', 'orange', 'green'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=20, color='purple', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
labels = df['Marital status'].value_counts().index
sizes = df['Marital status'].value_counts().values

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Marital Status Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Example data
years = df['Your current year of Study'].value_counts().index
counts = df['Your current year of Study'].value_counts().values

plt.figure(figsize=(8, 6))
plt.plot(years, counts, marker='o', linestyle='-', color='teal')
plt.title('Distribution of Students by Year of Study')
plt.xlabel('Year of Study')
plt.ylabel('Number of Students')
plt.show()
```
